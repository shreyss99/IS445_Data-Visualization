# Visualization 1

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sport_category = ['Aquatics & Diving', 'Athletics', 'Martial Arts', 'Cycling', 'Racquet / Bat Sports', 'Gymnastics', 
                  'Weightlifting & Powerlifting', 'Lawn Bowls', 'Triathlon', 'Netball & Basketball', 'Others']
```

```python
medals = [66,59,42,26,25,20,20,11,5,5,4]
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,12))

ax.pie(medals, labels = sport_category, autopct='%1.1f%%')
plt.legend(sport_category, loc = 'center')
plt.show()
```

# Visualization 2

```python
time = [1800, 1830, 1860, 1890, 1920, 1950, 1980, 2010, 2013, 2016, 2019, 2022]
human_population = [500, 750, 900, 1050, 1180, 1500, 1800, 4000, 5300, 5900, 6400, 7500]
species_extinction = [2, 5, 10, 20, 50, 100, 500, 1800, 2500, 4000, 5500, 7000]

fig, ax = plt.subplots(1,2,figsize=(16,5), sharex=True, sharey=True)

ax[0].plot(time, species_extinction, color='r')
plt.setp(ax[0:2], xlabel='Time')
plt.setp(ax[0:1], ylabel='Species Extinction (numbers)')

ax[1].plot(time, human_population, color='b')
plt.setp(ax[1:2], ylabel='Human population (millions)')

plt.show()
```

# Visualization 3

```python
import numpy as np

sugar = [0, 0, 1.2, 1.87, 3, 4, 4.36, 4.6, 4.91, 5, 5, 6, 6.55, 7.09, 7.1, 7.5, 7.65, 7.8, 8.04, 8.1, 8.6, 9.27, 10,
         10, 11.3, 14.5]
calories = [60, 85, 110, 93, 110, 100, 120, 108.9, 114.5, 100, 110, 120, 114.5, 115.3, 112.5, 123.5, 104.3, 128.6,
            99.3, 103.9, 106.9, 120, 130, 106.4, 103.7, 106]

fig, ax = plt.subplots(1,1,figsize=(8,5))
ax. scatter (sugar, calories)
plt.xlabel("Sugar Content")
plt.ylabel("Calories Consumed")

b, a = np.polyfit (sugar, calories, deg=1)
xseg = np. linspace(0,sugar)

plt.plot(xseg, a + b * xseg,color="r")
plt.show()
```
