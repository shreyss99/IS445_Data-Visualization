# Relation between Year Acquired and Year Constructed 

```python
data[['Year Constructed', 'Year Acquired']]
```

```python
# Arithmetic operation

data['Age at Acquisition'] = data['Year Acquired'] - data['Year Constructed']
# Add a new column 'Age at Acqusition' to the dataframe

data.head()
```

```python
data[['Age at Acquisition']]
```

```python
# We see there are some values which are negative and we need to remove them
# So we will filter out the data

# Filter by relationship

filter_1 = data['Year Acquired'] > 0 
# for selecting rows that the value in 'Year Acquired' greater than 0

filter_2 = data['Year Constructed'] > 0
# for selecting rows that the value in 'Year Constructed' greater than 0

filter_3 = data['Age at Acquisition'] >= 0
# for selecting rows that the value in 'Age at Acquisition' greater than or equal to 0


data_filtered = data.loc[(filter_1)&(filter_2)&(filter_3)].copy()
# .loc[] select rows based on the filter (or filters) applied to the column (or columns)

# .copy() is to copy the selected rows as a new dataframe

print(data_filtered.shape) # print n_rows, n_cols in the new dataframe (the selected one)
```

```python
# Let's plot the scatter plot of Year Constructed vs Year Acquired

fig, ax = plt.subplots(figsize=(15,15))
ax.scatter(data_filtered['Year Constructed'], data_filtered['Year Acquired'])

ax.set_xlabel("Year Constructed", fontsize=13)
ax.set_ylabel("Year Acquired", fontsize=13)
ax.set_title("Relationship between Year Constructed and Year Acquired", fontsize=13)
ax.set_xticks([i for i in range(1750, 2020, 10)])
ax.set_yticks([i for i in range(1750, 2020, 10)])
ax.set_xticklabels(ax.get_xticks(), rotation = 45)

plt.show()
```

- In this question, I want to find the relationship between the Year Constructed and Year Acquired variables. 
- The shortcoming of the question can be that there are a certain values as 0 for the years which do not make sense.    - So before plotting, we have performed a few filtering operations where we check when Year Acquired and Year Constructed both are >= 0 and also the difference between them, if negative, we remove those specific rows.
- Once done, we then use a scatter plot to plot these 2 variables because both of them are numerical variables and scatterplot seems the best plot.

- Another shortcoming that I feel with this approach is that eventhough we plot a scatterplot, the graph does not make much sense since we are just comparing 2 different year values. Only a thing that I can figure out is that when the Year Constructed is high the Year Acquired is same or greater because it makes sense that once the building has been constructed, it can only be acquired in that year or later years.
- This question can be done in a more meaningful way as to how many buildings had been acquired in a specific year interval as a function of the year Acquired.

# Total Square Footage as function of Congress District

```python
# Finding the sum of 'Square Footage' and group it by the 'Congress Dist'

df_Total_Square_Footage_Congress_Dist = data_filtered.groupby('Congress Dist')[['Square Footage']].sum()
```

- The 'Congress Dist' is a categorical variable and the 'Square Footage' is a numberical variable. 
- I found the unique values for Congress Dist variable and see that there are 19 values from 0 to 18. 
- I created a new variable where I am calculating the sum of Square Footage for each Congress Dist and group according to the Congress Dist. 

- I have then used a bar chart to plot the total Square Footage against the categorical Congress Dist and set the labels because the Congress Dist can be considered as a categorical variable and the area as numeric. 
As far as I think there is no shortcoming for this question and I have successfuly been able to provide an informative visualization.

# Average square footage per floor as a function of congressional district

```python
data_filtered[['Congress Dist', 'Square Footage', 'Total Floors']]
```

```python
# Displaying the unique values in 'Total Floors'

data_filtered['Total Floors'].unique()
```

```python
# Displaying the value counts in 'Total Floors'

data_filtered['Total Floors'].value_counts()
```

```python
# Finding the average of 'Square Footage' and group it by the 'Congress Dist' and 'Total Floors'

df_Average_Square_Footage_Per_Floor_Congress_Dist = data_filtered.groupby(
                                                    ['Congress Dist', 'Total Floors'])[['Square Footage']].mean()
```

```python
df_Average_Square_Footage_Per_Floor_Congress_Dist
```

```python
# Calculating the average area per floor for congress dist using groupby and unstack functions 
# to create a new dataframe
# unstack helps to get the third variable (Area) apart from Congress Dist and Floor number

df_Average_Square_Footage_Per_Floor_Congress_Dist = data_filtered.groupby(
    ['Congress Dist',"Total Floors"])['Square Footage'].mean().unstack('Total Floors').fillna(0)
```

```python
# Plotting a stacked bar chart 

fig, ax = plt.subplots(1,1, figsize=(18,15))
ax.set_xlabel("Congress District", fontsize=15)
ax.set_ylabel("Average Square Footage (in 1e7 sq units)", fontsize=15)
ax.set_title("Average Square Footage per Floor as function of Congress Dist", fontsize=15)
ax.set_xticks(fontsize=15)

df_Average_Square_Footage_Per_Floor_Congress_Dist.plot(kind='bar', stacked=True, ax=ax)
plt.legend(df_Average_Square_Footage_Per_Floor_Congress_Dist.index,
           bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, title='Departments')


plt.show()
```

- For this problem, we need to find the sum of all square footage areas for each congress district. 
- I have checked before and there are no null values in both 'Congress Dist', 'Square Footage' and 'Total Floors. 
- The 'Congress Dist' is a categorical variable and the 'Square Footage' and 'Total Floors' are numberical variables. - I found the unique values for Congress Dist and Total Floors variable. There are 19 values from 0 to 18 for Congress Dist and 24 values for Tota Floors. 
- I created a new variable where I am calculating the average of Square Footage per floor for each Congress Dist and group according to the Congress Dist and Total Floors. 
- In this case, to plot the graphs for each Congress District, I have used a stacked bar chart because I want to plot multiple floor areas for a single district.

- I have interpreted this question as finding the average of the total square footage for each floor that is present in the district. This can also be interpreted in another way that we find the sum of the total square footage for a district and divide it by the total number of floors in that district. So this is a shortcoming as the question is ambiguous. I found the first approach more meaningful and so I chose it.

# Square footage for the five most common departments as a function of year

```python
# Using the filtered data to group on 2 columns of Year Acquired and agency name.
# We then use the agg function using the sum for Square Footage and count for the Agency Name

df_all_dept = data_filtered.groupby(['Year Acquired', 'Agency Name']).agg({'Square Footage': 'sum',
                                                                           'Agency Name': 'count'})
df_all_dept
```

```python
df_renamed = df_all_dept.rename(columns={'Agency Name':'Dept Count', 'Square Footage':'Square Footage Sum'})
df_renamed
```

```python
df_Descending_Count = df_renamed.sort_values('Dept Count', ascending = False)
df_Descending_Count_Modified = df_Descending_Count.reset_index()
df_Descending_Count_Modified
```

```python
# Make a new dataframe
df_temporary = pd.DataFrame(columns = ['Year Acquired', 'Agency Name', 'Square Footage Sum'])

# Find unique names of departments
unique_depts = list(df_Descending_Count_Modified['Agency Name'].unique())

for i in unique_depts:
    df_department_names = df_Descending_Count_Modified[df_Descending_Count_Modified['Agency Name'] == i]
    df_temporary.loc[len(df_temporary.index)] = df_department_names.iloc[0]
```

```python
# Top 5 departments

df_top5_Departments_raw = df_temporary.head(5)
df_top5_Departments_raw
```

```python
# Final list of top 5 departments

df_top5_Departments = df_top5_Departments_raw.groupby(['Year Acquired','Agency Name'])['Square Footage Sum'].sum().unstack('Agency Name').fillna(0)
fig, ax = plt.subplots(1,1, figsize=(18,15))

df_top5_Departments.plot(kind='bar', stacked=True, ax=ax)

plt.xlabel('Year Acquired', fontsize=15)
plt.ylabel('Square Footage Sum in (1e7 sq units)', fontsize=15)
plt.title('Total Square Footage for 5 most common departments as a function of Year Acquired', fontsize=15)
plt.xticks(fontsize=13, rotation=45)
plt.legend(title='Departments', loc='upper left', bbox_to_anchor=(1.05, 1))

plt.show()
```

- In this question, we need to find the total square footage for top 5 common departments as a function of time.
- The approach I have taken is that I have firstly performed a groupby taking 2 columns - Agency Name and the Year Acquired. Along with it, I am performing the sum for square footage and count of the Agency Name to get the top 5 departments.
- I have then sorted the data obtained in descending order based on the total square footage sum and renamed columns.
- Nextly, I have created a new dataframe and then I traverse the list of unique departments. This loop will be used for inserting the useful rows in the new dataframe for getting the final list.
- From the dataframe, I choose the top 5 values for plotting purpose.
- Then I have used a stacked bar chart for plotting purpose because we have used the aggregation and unstack functions.
